# Pylot API vs Object-Oriented(OO) API
#pylot API - Quickly
#OO API -> Advanced

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),color='blue',linestyle='dashed');
plt.xlabel('Bien x');
plt.ylabel('Sin(X)');
# plt.xlim(0,4);
# plt.ylim(-0.75,0.6);
```

```python
#ve chung do thi
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),color='blue',linestyle='dashed',label='sin(x)');
plt.plot(x,np.cos(x),color='red',label='cos(x)')
plt.xlabel('Bien x');
plt.ylabel('Sin(X)');
# name graphic
plt.title('Sin(x) and Cos(x)',color='green');
# plt.xlim(0,4);
# plt.ylim(-0.75,0.6);
plt.axis([0, 4, -0.7, 0.6]);#xmin,xmax,ymin,ymax tuong tu nhu xlim
plt.legend(); 
```

# Bar
* vertical

```python
soft_drink_prices = {'coke':10,
                    'pessi':12,
                    'sprite':8}
fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax.set(title='Bach Hoa Xanh Soft',ylabel = 'Price($)');
```

```python
fig,ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),
        list(soft_drink_prices.values()));
```

# Histogram

```python
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
plt.hist(student_height,bins=30);
```

# Subplots

```python
#option1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                     ncols = 2,
                     figsize = (10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(list(soft_drink_prices.keys()),
        list(soft_drink_prices.values()));
ax4 = plt.hist(student_height,bins=30);
```

```python
# option2
fig,ax = plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(20,10)
                    )
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(list(soft_drink_prices.keys()),
        list(soft_drink_prices.values()));
ax[1,1].hist(student_height,bins=30);
```

# Example 

```python
cities = pd.read_csv('california_cities.csv')
```

```python
cities.head()
# cities['area_total_km2'].sort_values(ascending= False)
```

# note 
### latd v  , longd kinh 
### population -dn s
### area_total_km2 - dien tich

```python
# extract latd , longd
lat,lon = cities['latd'],cities['longd']
# lay them thong tin nhu dan so
polulation,area = cities['population_total'],cities['area_total_km2']
# plt.figure(figsize(8,10))
plt.style.use('seaborn')
# plot using pylot API
plt.scatter(lon,lat,c=np.log10(polulation),
            cmap='viridis',s=area,
           linewidths=0,alpha=0.5);
# trang tri
plt.axis('equal');
plt.xlabel('longtitude')
plt.ylabel('latitude');
# thanh tab3
plt.colorbar(label='log$_{10}&(population)');
plt.clim(3,7);


#creat a lengend for cites size
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,c='k',alpha=0.4,
                
                label=str(area) + 'km$^2$')
plt.legend(labelspacing = 1,title='City Area') #scatterpoints=1


plt.title('California Cites: Population and Area Distribution');
```
